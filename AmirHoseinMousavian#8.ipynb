# DataSet

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Data = pd.read_csv('Market.csv') # calling the dataset
df = pd.DataFrame(Data) # making a dataframe
df
```

```python
df1 = df[df['Index'] == 'NYA'] # selecting the wanted part
```

```python
df1
```

```python
df1.describe() # Describing data gives us great insights about data especialy about missing values. 
```

```python
df1.isnull().sum() # this shows us how many missing values we have.
```

# Open and Adj Close

```python
plt.figure(figsize=(15,5)) # setting the plot size
plt.scatter(df1['Open'],df1['Adj Close'], c='purple', alpha=0.5) # setting x and y in data, colors and mods
plt.title("Open & Adj Close", loc='left', c='blue') # setting the title of the plot
plt.xlabel("Open", c='r') # setting the x axis lable
plt.ylabel("Adj Close", c='red') # setting the y lable
plt.grid(linewidth = 0.2, color='black') # setting the grid for plot, color and width
plt.show() # showing the plot
```
